```python
imgtype1 = 'image_2'
imgtype2 = 'image_3'
```

```python
datadir = '/scratch/datasets/kitti2015/training/'
```

```python
paths1 = sorted(glob.glob(datadir+'obj_map'+'/*.png'),key=os.path.getmtime)
for i, pth1 in enumerate(paths1):
    arr1 = (imread(pth1) -128)/128
    print(arr1)
    plt.imshow((arr1+1)/2)
    plt.show()
    if i==3: break
```

```python
paths1 = sorted(glob.glob(datadir+imgtype1+'/*.png'),key=os.path.getmtime)
paths2 = sorted(glob.glob(datadir+imgtype2+'/*.png'),key=os.path.getmtime)
for i, (pth1, pth2) in enumerate(zip(paths1,paths2)):
    arr1 = (imread(pth1) -128)/128
    arr2 = (imread(pth2) -128)/128
    
    diff = arr1 - arr2
    #print(diff)
    plt.imshow((arr1+1)/2)
    plt.show()
    plt.imshow((arr2+1)/2)
    plt.show()
    plt.imshow((diff-np.min(diff))/np.max(diff))
    plt.show()
    #display(Image(filename=pth1)-Image(filename=pth2))
    if i==1: break
        
    
# import matplotlib.pyplot as plt

# im = cv2.imread('image.jpg')
# im_resized = cv2.resize(im, (224, 224), interpolation=cv2.INTER_LINEAR)

# plt.imshow(cv2.cvtColor(im_resized, cv2.COLOR_BGR2RGB))
# plt.show()
```
