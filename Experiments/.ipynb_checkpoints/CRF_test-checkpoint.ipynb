```python
def rbf(center,sigma):
    sigma = np.array(sigma)[None,None,...]
    center = np.array(center)[None,None,...]
    ic, jc = center[:,:,0,...], center[:,:,1,...]
    return lambda i,j: np.exp(-((i[...,None]-ic)**2 + (j[...,None]-jc)**2)/(2*sigma**2)).squeeze()
```

```python
def to_matrix(A):
    s = A.shape
    return A.reshape((s[0]*s[1],s[2]*s[3]))

def gk2d(xa,xb,sigma):
    diff = xa[...,:,:,None,None]-xb[...,None,None,:,:]
    #print(diff.shape)
    dist = to_matrix(np.linalg.norm(diff,axis=0))
    return -np.log(dist/sigma+ np.eye(*dist.shape))/(2*np.pi)+np.log(sigma)*np.eye(*dist.shape)/(2*np.pi)

def rbf2d(xa,xb,sigma):
    diff = xa[...,:,:,None,None]-xb[...,None,None,:,:]
    #print(diff.shape)
    dist = to_matrix(np.linalg.norm(diff,axis=0))
    return np.exp(-(dist/sigma)**2)

def lap2d(xa,xb,sigma):
    diff = xa[...,:,:,None,None]-xb[...,None,None,:,:]
    #print(diff.shape)
    dist = to_matrix(np.linalg.norm(diff,axis=0))
    return np.exp(-dist/sigma)
```

```python
eps=1e-8
nx = ny = 30
x = np.linspace(-30,30,nx)
y = np.linspace(-30,30,ny)
xa = np.array(np.meshgrid(x,y))

sig = 20
K = gk2d(xa,xa,sig)
eigs,V = np.linalg.eigh(K)
#print(eigs)
eigs[eigs<eps]=eps
K_new = V@np.diag(eigs)@V.T

K_rbf = rbf2d(xa,xa,sig) + eps*np.eye(*(nx*ny,nx*ny))
K_lap = lap2d(xa,xa,sig) + eps*np.eye(*(nx*ny,nx*ny))


fig = plt.figure(figsize=(10,20))
plt.subplot(131)
plt.imshow(K_new)
plt.subplot(132)
plt.imshow(K_lap)
plt.subplot(133)
plt.imshow(K_rbf)
#plt.subplot_tool()
plt.show()
```

```python
laplacian_2d = np.array([[0,-1,0],
                         [-1,4,-1],
                         [0,-1,0]])
img_K_lap = K_new.reshape((30,30,30,30))
LK = to_matrix(sp.ndimage.filters.convolve(img_K_lap, laplacian_2d[...,None,None], mode='reflect'))
Lsk = LK#+K_new/sig
plt.imshow(Lsk)
plt.colorbar()
```

```python
Lnew = np.linalg.cholesky(K_new)
Llap = np.linalg.cholesky(K_lap)
Lrbf = np.linalg.cholesky(K_rbf)
```

```python
ndraws = 5
z = np.random.normal(size = (nx*ny,ndraws))
draws = (Lnew@z).reshape((nx,ny,ndraws))
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
nderivs = 3
fig = plt.figure(figsize=(40,40))


for k,L in enumerate([Lnew,Llap,Lrbf]):
    z = np.random.normal(size = (nx*ny))
    draw = (L@z).reshape((nx,ny))
    img = draw
    for i in range(nderivs):
        plt.subplot(3*100+nderivs*10+k*3+i+1)
        ax = plt.gca()
        im = ax.imshow(img,interpolation=None)
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="10%", pad=0.05)
        plt.colorbar(im, cax=cax)
        
        img = sp.signal.convolve(img,laplacian_2d,mode='valid')
        
plt.show()
```

```python
for i in range(ndraws):
    deriv2 = sp.signal.convolve(draws[:,:,i],laplacian_2d,mode='valid')
    deriv4 = sp.signal.convolve(deriv2,laplacian_2d,mode='valid')
    plt.imshow(deriv4,interpolation=None)
    plt.colorbar()
    plt.show()
```
